```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

# use webdriver to find elements:
from selenium.webdriver.common.by import By 
```

```python
# for creating dataframe and writing documents
!pip install pandas
import pandas as pd
```

```python
# for parsing html document
from bs4 import BeautifulSoup
import lxml
```

```python
#for simulating automoted actions
from selenium.webdriver import Chrome, ActionChains, Keys
```

```python
# add time delays to make sure web loading completely
from time import sleep
```

```python
#call two browsers to do the loop crawl and create a nested loop:
#the first browser is responsible for looping through the href elements of the 100 product display pages
browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
browser.get('https://www.artsy.net/collection/contemporary?page=13&sort=sold%2C-has_price%2C-prices')

#the second browser is responsible for looping trough detail pages's URL retrieved from the first browser
browser2 = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# tells WebDriver to wait when trying to find any elements that are not immediately available. 
browser.implicitly_wait(3)
sleep(2)
```

```python
# create list for storing scrabbed data
data_list = []  
```

```python
#In the outer loop, find the <href> elements in the 100 pages 
for i in range(1,101):
    soup = BeautifulSoup(browser.page_source, 'lxml')
    div_list = soup.find_all('div', relay='[object Object]')

    # print <href> and create the URL list of each artwork's detail page in order:
    for div in div_list:
        link = 'https://www.artsy.net' + div.find('a').get('href')
    # In the inner loop,accessing the URL of the details page in order        
        browser2.get(link)
        sleep(1)
        browser2.implicitly_wait(30)
        
        # simulate scroll-down action to load each page completely
        actions = ActionChains(browser2)
        for j in range(5):
            actions.send_keys(Keys.PAGE_DOWN).perform()
            sleep(3)
        # parsing the html of what the browser2 get
        soup = BeautifulSoup(browser2.page_source, 'lxml')

# Using beautiful soup to find the element
# if some elements are not included in the html documents, output 'NA' and run the next step       
        try:
            artist = soup.find('a',
                           class_='RouterLink__RouterAwareLink-sc-1nwbtp5-0 ekDWRq ArtworkSidebar2Artists__StyledArtistLink-gyavil-0 iKbnpN').text
        except:
            artist = 'NA'
        try:
            artwork = soup.find('h1', class_='Box-sc-15se88d-0 Text-sc-18gcpao-0 caIGcn fiPLKL').text
            # split artwork text by separator comma
            artwork_name=artwork.split(',')[0]
            artwork_date=artwork.split(',')[1]
        except:
            artwork = 'NA'
            artwork_name='NA'
            artwork_date='NA'  
        try:
            price = soup.find('div', class_='Box-sc-15se88d-0 Text-sc-18gcpao-0 eXbAnU lcTzuI').text
        except:
            price = 'NA'  
        
        try:
            medium = soup.find('div', class_='Box-sc-15se88d-0 Text-sc-18gcpao-0 HTML__Container-sc-1im40xc-0 caIGcn bJVMLf fxdlkC').text
        except:
            medium = 'NA'
        
        try:
            w = soup.find_all('div', class_='Box-sc-15se88d-0 Flex-cw39ct-0 edNsZM') 
        except:
            w = 'NA'
        # find the text elements in the first div class of w
        try:
            gallery = w[0].find('div').text
        except:
            gallery = 'NA'

        # find the second text elements in the second div class of w
        try:
            m = w[1].find_all('div')[1].text
        except:
            m = 'NA'
        try:
            place = m.split(',')[0]
        except:
            place = 'NA'
        try:
            date = m.split(',')[1]
        except:
            date = 'NA'
               
            

# creating the dictionary to store data        
        dic = {
            'artist': artist,
            'artwork': artwork_name,
            'artwork_date':artwork_date,
            'gallery': gallery,
            'price': price,
            'place': place,
            'author_date_of_birth': date,
            'medium':medium
        }
# output the data we scrabbed from the html
        print(dic)
# append data to the data_list we created at the beginning        
        data_list.append(dic)
# use css selector find and simulates a click on the 'Next' element        
    pa = browser.find_element(By.CSS_SELECTOR,'[data-testid="next"]')
    pa.click()
    sleep(3)    
         
```

```python
# output the amount of dictionaries and check whether all the information have been included
len(data_list)
```

```python
# create the dataframe and define the column name
artsy_data=pd.DataFrame(data_list,index=range(1,len(data_list)+1),columns=['artist','author_date_of_birth','artwork','artwork_date','price','medium','gallery','place'],dtype=object)
artsy_data
```

```python
# output documents as xlsx
artsy_data.to_excel('artworks.xlsx')
```
